```python
# train_columns = ["geo_place_name", "geo_type_name", "Ranking of Average Benzene Concentrations", "Ranking of Average Formaldehyde Concentrations", "Ranking of Fine Particulate Matter (PM2.5)", "Ranking of Nitrogen Dioxide (NO2)", "Ranking of Ozone (O3)", "Ranking of Sulfur Dioxide (SO2)", "Final Ranking"]

train_columns = ["geo_place_name", "geo_type_name", "Air Toxics Concentrations- Average Benzene Concentrations", "Air Toxics Concentrations- Average Formaldehyde Concentrations", "Fine Particulate Matter (PM2.5)", "Nitrogen Dioxide (NO2)", "Ozone (O3)", "Sulfur Dioxide (SO2)", "Final Ranking"]
```

```python
# from sklearn.preprocessing import LabelEncoder

# le = LabelEncoder()
# for column in train_columns:
#     le.fit(df[column])
#     df[column] = le.transform(df[column])

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for column in ["geo_place_name", "geo_type_name", "Final Ranking"]:
    le.fit(df[column])
    df[column] = le.transform(df[column])

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df[train_columns[:-1]], df[train_columns[-1]], test_size=0.2)

from sklearn.preprocessing import StandardScaler

scl = StandardScaler()

X_train = pd.DataFrame(scl.fit_transform(X_train), columns = X_train.columns)

X_test = pd.DataFrame(scl.transform(X_test), columns = X_test.columns)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_validate
from sklearn import metrics
from sklearn.metrics import f1_score
from numpy import mean

lrm = LogisticRegression()
dtm = DecisionTreeClassifier(random_state=0, max_depth=5)
gnbm = GaussianNB()

models = [lrm, dtm, gnbm]
score_lst = []
mean_score_lst = []

for m in models:
    # scores = cross_validate(m, X_train, y_train, scoring = ['accuracy','precision','recall','f1'], cv=5)
    # scores = cross_validate(m, X_train, y_train, scoring = ['accuracy','precision_macro','recall_macro','f1_macro'], cv=5)
    scores = cross_validate(m, X_train, y_train, scoring = ['accuracy','precision_weighted','recall_weighted','f1_weighted'], cv=5)
    scores = pd.DataFrame(scores)
    score_lst.append(scores.drop(['fit_time', 'score_time'], axis = 1))
    mean_score = pd.DataFrame(scores.mean()).drop(index = ['fit_time', 'score_time'])
    mean_score_lst.append(mean_score)
```

```python
display('Logistic Regression performance on 5 folds is:', '\n')
score_lst[0]
```

```python
display('Logistic Regression Mean performance on 5 folds is:', '\n')
mean_score_lst[0]
```

```python
display('Decision Tree performance on 5 folds is:', '\n')
score_lst[1]
```

```python
display('Decision Tree Mean performance on 5 folds is:', '\n')
mean_score_lst[1]
```

```python
display('Naive Bayes performance on 5 folds is:', '\n')
score_lst[2]
```

```python
display('Naive Bayes Mean performance on 5 folds is:', '\n')
mean_score_lst[2]
```

```python
dtm.fit(X_train, y_train)
y_pred = dtm.predict(X_test)
report = metrics.classification_report(y_test, y_pred, output_dict=True)
df_classification_report = pd.DataFrame(report).transpose()
display(report)
display(df_classification_report)
unique, counts = np.unique(y_pred, return_counts=True)

print(np.asarray((unique, counts)).T)
```
