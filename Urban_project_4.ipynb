```python
# from sklearn.preprocessing import LabelEncoder

# le = LabelEncoder()
# for column in train_columns:
#     le.fit(df[column])
#     df[column] = le.transform(df[column])

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
frames = [trainset, testset]
df_traintest = pd.concat(frames)
display(df_traintest)
le_name_mapping = {}
for column in ["geo_place_name", "geo_type_name","geo_borough_name", "Final Ranking"]:
    # le.fit(trainset[column])
    # trainset[column] = le.transform(trainset[column])
    # le.fit(testset[column])
    # testset[column] = le.transform(testset[column])

    # le.fit(df[train_columns][column])
    # df[train_columns][column] = le.transform(df[train_columns][column])

    le.fit(df_traintest[column])
    le_name_mapping[column] = dict(zip(le.classes_, le.transform(le.classes_)))
    # print(le_name_mapping)
    df_traintest[column] = le.transform(df_traintest[column])

print(le_name_mapping["geo_borough_name"])
# trainset_bronx = df[train_columns].loc[(df["geo_borough_name"] != "Bronx") ]
# testset_bronx = df[train_columns].loc[df["geo_borough_name"] == "Bronx"]

trainset_bronx = df_traintest.loc[(df_traintest["geo_borough_name"] != le_name_mapping["geo_borough_name"]["Bronx"]) ]
testset_bronx = df_traintest.loc[(df_traintest["geo_borough_name"] == le_name_mapping["geo_borough_name"]["Bronx"]) ]

trainset = trainset_bronx
testset = testset_bronx

X_train = trainset.drop(["Final Ranking"], axis=1)
y_train = trainset["Final Ranking"]

X_test = testset.drop(["Final Ranking"], axis=1)
y_test = testset["Final Ranking"]

print("X_train after encoding")
display(X_train)
print("X_test after encoding")
display(X_test)
print("y_train after encoding")
display(y_train)
print("y_test after encoding")
display(y_test)

from sklearn.preprocessing import StandardScaler

scl = StandardScaler()

X_train = pd.DataFrame(scl.fit_transform(X_train), columns = X_train.columns)
print("X_train after scaling")
display(X_train)

X_test = pd.DataFrame(scl.transform(X_test), columns = X_test.columns)
print("X_test after encoding")
display(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_validate
from sklearn import metrics
from sklearn.metrics import f1_score
from numpy import mean

lrm = LogisticRegression()
dtm = DecisionTreeClassifier(random_state=0, max_depth=5)
gnbm = GaussianNB()

models = [lrm, dtm, gnbm]
score_lst = []
mean_score_lst = []

for m in models:
    # scores = cross_validate(m, X_train, y_train, scoring = ['accuracy','precision','recall','f1'], cv=5)
    #scores = cross_validate(m, X_train, y_train, scoring = ['accuracy','precision_macro','recall_macro','f1_macro'], cv=5)
    scores = cross_validate(m, X_train, y_train, scoring = ['accuracy','precision_weighted','recall_weighted','f1_weighted'], cv=5)
    scores = pd.DataFrame(scores)
    score_lst.append(scores.drop(['fit_time', 'score_time'], axis = 1))
    mean_score = pd.DataFrame(scores.mean()).drop(index = ['fit_time', 'score_time'])
    mean_score_lst.append(mean_score)
```

```python
display('Logistic Regression Mean performance on 5 folds is:', '\n')
mean_score_lst[0]
```

```python
display('Naive Bayes performance on 5 folds is:', '\n')
score_lst[2]
```

```python
lrm.fit(X_train, y_train)
y_pred = lrm.predict(X_test)
print(type(y_pred))
display(set(y_pred))
display(y_pred[:5])
report = metrics.classification_report(y_test, y_pred, output_dict=True)
df_classification_report = pd.DataFrame(report).transpose()

unique, counts = np.unique(y_pred, return_counts=True)

print(np.asarray((unique, counts)).T)
display(report)
display(df_classification_report)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print('Accuracy score: {}'.format(accuracy_score(y_test, y_pred)))
print('Precision score: {}'.format(precision_score(y_test, y_pred)))
print('Recall score: {}'.format(recall_score(y_test, y_pred)))
print('F1 score: {}'.format(f1_score(y_test, y_pred)))

print(metrics.classification_report(y_test, y_pred))
```
